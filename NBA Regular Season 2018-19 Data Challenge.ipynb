# What was the average age of player in the league?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re

nba_stats = pd.read_csv('nbastats2018-2019.csv')
nba_stats.head()

```

# What player scored the most points?

```python
nba_stats_points = nba_stats.groupby(['Name', 'Points']).head()
nba_stats_points.sort_values('Points', ascending=False)

# James Harden had the most points
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
nba_stats_blocks = nba_stats.groupby('Name')['Blocks'].max().sort_values(ascending=False)
nba_stats_blocks
```

# What player had the best 3-pt percentage? 

```python
nba_stats_3pt = nba_stats.sort_values(by='FG3%', ascending=False).filter(['Name', 'FG3%'])
nba_stats_3pt.head(5)

```

# Who played the most minutes during the season

```python
nba_stats_MP = nba_stats.sort_values(by='MP', ascending=False).filter(['Name', 'MP'])
nba_stats_MP.head(5)
```

# What player given their player effiecency rating was the clutchest during the season?

```python
nba_stats_win_percentage = nba_stats.groupby('Name')['PER'].max().sort_values(ascending=False)
nba_stats_win_percentage

# Zhou Qi had the highest PER.
```

# What team had the youngest roster?

```python
nba_stats_youngest = nba_stats.groupby('Team')['Age'].mean().sort_values()

nba_stats_youngest
```

# Who is the highest paid player during the seasion?

```python
# nba_stats_salary = nba_stats.groupby('Name')['Salary'].max().sort_values()
# nba_stats_salary

# I had to convert the Salary data to numeric as it was reading for this code as a string
nba_stats['Salary'] = pd.to_numeric(nba_stats['Salary'], errors='coerce')

nba_stats_salary = nba_stats.groupby('Name')['Salary'].max().sort_values(ascending=False)

nba_stats_salary.head(15)
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
nba_stats_free_throw = nba_stats.groupby(['Name', 'FT%', 'FTA']).min()

nba_stats_free_throw.sort_values('FT%', ascending=True)
```
